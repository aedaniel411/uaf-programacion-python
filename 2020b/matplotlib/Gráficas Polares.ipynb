# Gr√°ficas polares

```python
# importamos modulos
import numpy as np
import matplotlib.pyplot as plt

# definimos datos
theta = np.arange(0., 2., 1./180.)*np.pi

# Definimos gr√°fico
plt.figure(10)
plt.polar(3*theta, theta/6)
plt.polar(theta, np.cos(6*theta))
plt.polar(theta, [1.2]*len(theta))

# Guardar y mostrar gr√°fico
plt.savefig('polar01.png')
plt.show()
```

```python
# definimos datos
nums = np.arange(0,10,0.1)

# definimos gr√°fico
plt.figure(11)
plt.subplot(111, projection='polar')
plt.plot(nums, nums/10.)
plt.plot(nums, np.sin(nums))
plt.plot(nums, np.cos(nums))

# Mostramos gr√°fico
plt.show()
```
