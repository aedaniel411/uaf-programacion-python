```python
import matplotlib.pyplot as plt          

plt.subplot(211)             
plt.plot([1, 2, 3])

plt.subplot(212)             
plt.plot([4, 5, 6])

plt.show()
```

```python
import matplotlib.pyplot as plt    
plt.figure(2)

plt.subplot(2, 1, 1)
plt.plot([1, 2, 3])

plt.subplot(2, 2, 3)
plt.plot([4, 5, 6])

plt.subplot(2, 2, 4)
plt.plot([7, 8, 9])
plt.show()
```
