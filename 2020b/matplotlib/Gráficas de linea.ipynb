## Gr√°ficas de linea

```python
import matplotlib.pyplot as plt
import numpy as np

#Definimos datos
t = np.arange(0., 5., 0.2)
#t = np.linspace(0., 5., 25)

#Definimos gr√°fica
plt.plot(t, t)
plt.plot(t, t**2) 
plt.plot(t, t**3)

#Mostramos gr√°fica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Definimos datos
#t = np.arange(0., 5., 0.2)
t = np.linspace(0., 5., 25)

#Definimos gr√°fica
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')

#Mostramos gr√°fica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

nums = np.arange(0,10,0.1)

plt.plot(nums, np.sin(nums), label='1')
plt.plot(1+nums, np.sin(nums), label='2')
plt.plot(2+nums, np.sin(nums), label='3')
plt.plot(3+nums, np.sin(nums), label='4')
plt.plot(4+nums, np.sin(nums), label='5')

#plt.legend(loc='best')
#plt.legend(loc=10)
plt.legend(loc='center right')
plt.show()
```
