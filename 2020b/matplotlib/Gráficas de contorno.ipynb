# Gráficas de contorno

```python
# importar modulos
import numpy as np
import matplotlib.pyplot as plt

# definir datos
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)

# Definimos gráfico
plt.figure(10)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)

# Muestra barra de colores
fig.colorbar(cp)

# Definimos titulo de gráfico y de ejes
ax.set_title('Gráfica de contorno')
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')

# Mostramos gráfico
plt.show()
```

```python
# Definimos datos 
x = np.sort(np.random.randn(25))
y = np.sort(np.random.randn(25))

mat1, mat2 = np.meshgrid(x, y)
mat = np.sqrt( mat1**2 + mat2 **2)

# Definimos gráfico
plt.figure(20)
plt.matshow(mat)

plt.colorbar()

#Motramos grafico
plt.show()
```

```python
# Definimos datos 
x = np.sort(np.random.randn(25))
y = np.sort(np.random.randn(25))

mat1, mat2 = np.meshgrid(x, y)
mat = np.sqrt( mat1**2 + mat2 **2)

# Definimos gráfico
plt.figure(20)
plt.matshow(mat) 
plt.contour(np.arange(25), np.arange(25), mat, 10, colors = 'k')

#Motramos grafico
plt.show()
```

```python
# Definimos datos

# posiciones X de nuestra red de medidas 
x = np.random.rand(20)
# posiciones Y de nuestra red de medidas
y = np.random.rand(20) 
# valores de Temperatura (oK) en las posiciones (X, Y) 
t = np.random.rand(20)*3000

# Definimos gráfico
plt.figure(30)
# Pintamos las triangulaciones con contornos de color 
plt.tricontourf(x, y, t)
# Pintamos las líneas de contorno en color negro 
plt.tricontour(x, y, t, colors = 'k')
plt.scatter(x, y) # Pintamos la posición de las estaciones de medida.

#Mostramos gráfico
plt.show()
```
