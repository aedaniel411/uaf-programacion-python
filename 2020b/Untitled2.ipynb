```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def funcion(x,a,b):
    return a / x**b

archivo = np.loadtxt('datos01.txt')

res, mat = curve_fit(funcion, archivo[:,0], archivo[:,1] )

x = np.linspace(min(archivo[:,0]), max(archivo[:,0]) )
y = funcion (x, res[0],res[1])

plt.scatter(archivo[:,0], archivo[:,1])
plt.plot(x, y, color='red')

plt.show()
```
