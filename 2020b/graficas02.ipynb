```python
#importar modulos
import numpy as np
import matplotlib.pyplot as plt

# definir datos
x = np.linspace(0,10,100)
y = x ** 2
z = x ** 3

# definir grafico
plt.figure(100)
plt.plot (x,y, label=r'$\alpha$')
plt.plot (x,z, label=r'$\beta$')

plt.legend()
plt.xlabel(r'$\alpha$')
plt.ylabel('Eje Y')

plt.show()
```

```python
plt.figure(90)
plt.axis([0,9,0,18])

arrstyles = ['-', '->', '-[', '<-', '<->', 'fancy', 'simple', 'wedge']
for i, style in enumerate(arrstyles):
    plt.annotate(style, xytext=(1, 2+2*i), xy=(4, 1+2*i), arrowprops=dict(arrowstyle=style))

connstyles=["arc", "arc,angleA=10,armA=30,rad=15", "arc3,rad=.2", "arc3,rad=-.2", 
            "angle", "angle3"]
for i, style in enumerate(connstyles):
    plt.annotate("", xytext=(6, 2+2*i), xy=(8, 1+2*i), arrowprops=dict(arrowstyle='->', 
                                                                       connectionstyle=style))

plt.show()
```
