## Ejemplo 1

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
x1 = np.random.randint(0, 100, 10)
y1 = np.random.randint(0, 100, 10)

x2 = np.random.randint(0, 100, 10)
y2 = np.random.randint(0, 100, 10)

#Definir características del gráfico
plt.scatter (x1, y1, label='Datos 1', color='red' )
plt.scatter (x2, y2, label='Datos 2', color='purple')

#Definir titulo y nombre de ejes
plt.title('Gráfico de dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar leyenda y gráfico
plt.legend()
plt.show()
```
