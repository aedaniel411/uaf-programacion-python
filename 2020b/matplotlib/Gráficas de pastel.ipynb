## Ejemplo 1

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
datos = np.random.randint(0, 100, 5)

#Definir características del gráfico
plt.pie (datos)

#Definir titulo
plt.title('Gráfico de Pastel')

#Mostrar gráfico
plt.show()
```

## Ejemplo 2

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
datos = np.random.randint(0, 100, 5)
personas = ['Ana', 'Andrea', 'Alejandra', 'Marta', 'Teresa']


#Definir características del gráfico
plt.pie (datos, labels=personas)

#Definir titulo y nombre de ejes
plt.title('Gráfico de Pastel')

#Mostrar gráfico
plt.show()
```

## Ejemplo 3

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
datos = np.random.randint(0, 100, 5)
personas = ['Ana', 'Andrea', 'Alejandra', 'Marta', 'Teresa']


#Definir características del gráfico
plt.pie (datos, labels=personas, shadow=True, autopct='%d%%')

#Definir titulo y nombre de ejes
plt.title('Gráfico de Pastel')

#Mostrar gráfico
plt.show()
```

## Ejemplo 4

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
datos = np.random.randint(0, 100, 5)
personas = ['Ana', 'Andrea', 'Alejandra', 'Marta', 'Teresa']


#Definir características del gráfico
plt.pie (datos, labels=personas, shadow=True, autopct='%.2f%%')

#Definir titulo y nombre de ejes
plt.title('Gráfico de Pastel')

#Mostrar gráfico
plt.show()
```
