## Ejemplo 1

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
a = np.random.randint(0, 100, 20)
bins = np.linspace(0, 100, 10)


#Definir características del gráfico
plt.hist (a, bins)

#Definir titulo y nombre de ejes
plt.title('Gráfico de Histograma')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar gráfico
plt.show()
```

## Ejemplo 2

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
a = np.random.randint(0, 100, 20)
bins = np.linspace(0, 100, 10)


#Definir características del gráfico
plt.hist (a, bins, rwidth=.8)

#Definir titulo y nombre de ejes
plt.title('Gráfico de Histograma')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar gráfico
plt.show()
```

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
a = np.random.randint(0, 100, 20)
bins = np.linspace(0, 100, 10)


#Definir características del gráfico
plt.hist (a, bins, rwidth=.8, color='lightgreen')

#Definir titulo y nombre de ejes
plt.title('Gráfico de Histograma')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar gráfico
plt.show()
```
