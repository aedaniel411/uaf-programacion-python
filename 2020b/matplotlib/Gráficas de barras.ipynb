## Gr√°ficas de barras

## Ejemplo 1

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
x1 = range(10)

y1 = np.random.randint(0, 100, 10)
y2 = np.random.randint(0, 100, 10)

#Definir caracter√≠sticas del gr√°fico
plt.bar (x1, y1, label='Datos 1', color='lightblue' )
plt.bar (x1, y2, label='Datos 2', color='orange')

#Definir titulo y nombre de ejes
plt.title('Gr√°fico de barras')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar leyenda y gr√°fico
plt.legend()
plt.show()
```

## Ejemplo 2

```python
# Importamos modulos
import matplotlib.pyplot as plt
import numpy as np 

# Definir Datos
x1 = np.arange(10)
y1 = np.random.randint(0, 100, 10)
x2 = x1 + .4
y2 = np.random.randint(0, 100, 10)

#Definir caracter√≠sticas del gr√°fico
plt.bar (x1, y1, label='Datos 1', color='lightblue', width=0.4)
plt.bar (x2, y2, label='Datos 2', color='orange', width=0.4)

#Definir titulo y nombre de ejes
plt.title('Gr√°fico de barras')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Mostrar leyenda y gr√°fico
plt.legend()
plt.show()
```
